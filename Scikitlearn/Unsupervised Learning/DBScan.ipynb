<h1>DBSCAN</h1> Density based clustering of applicatiions with noise (DBSCAN) uses the density of points in a region to form clusters. It has two main parameters:"epilson" and "min samples" using which it classifies each points as a core point, reachable point or noise point(outlier).

```python
from sklearn.cluster import DBSCAN
```

```python
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
sns.set_style("darkgrid")
%matplotlib inline
plt.rcParams["figure.figsize"] = (10, 11)
plt.rcParams["font.size"] = 11
```

```python
iris_df = sns.load_dataset("iris")
X =iris_df.select_dtypes(np.number)
```

```python
model = DBSCAN(eps=0.5, min_samples=10).fit(X)
```

```python
model.labels_
```

```python
sns.scatterplot(X, x="sepal_length", y="petal_length", hue=model.labels_)
```
