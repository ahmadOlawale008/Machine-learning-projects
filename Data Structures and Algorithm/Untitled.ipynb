```python
class Node:
    def __init__(self, data=None, next=None, prev=None):
        self.data = data
        self.next = next
        self.prev = prev
class LList:
    def __init__(self):
        self.head = None
    def insert_at_beginning(self, data):
        if self.head is None:
            self.head = Node(data, self.head, None)
        else:
            node = Node(data, self.head, None)
            self.head.prev = node
            self.head = node
    def insert_at_end(self, data):
        curNode = self.head
        if self.head is None:
            self.head = Node(data, None, None)
            return
        while curNode.next:
            curNode = curNode.next
        curNode.next = Node(data, None, curNode)
    def get_last_item_node(self):
        curNode = self.head
        while curNode.next:
            curNode = curNode.next
        return curNode
    def get_last_item(self) -> int:
        return self.get_element_by_index(self.len() - 1)
    def insert_multiple_values(self, data_lists):
        self.head = None
        for data in data_lists:
            self.insert_at_end(data)
    def len(self):
        total_length = 1
        curNode = self.head
        if curNode is None:
            print(0)
            return 0
        while curNode.next:
            total_length += 1
            curNode = curNode.next
        return total_length
    def remove_element_at(self, index):
        total_len = self.len()
        counter = 0
        curNode = self.head
        if index >= total_len or index < 0:
            raise Exception("Linked list index out of range")
            return
        if index == 0:
            self.head = self.head.next
            return
        while curNode.next:
            if counter == index - 1:
                if curNode.next.next:
                    curNode.next = curNode.next.next
                    counter += 1
                else:
                    curNode.next = None
                    break
                continue
            curNode = curNode.next
            counter += 1
    def get_element_position(self, data):
        curNode = self.head
        if self.head and self.head.data == data:
            return 0
        counter = 1
        found = False
        while curNode.next:
            if curNode.next.data == data:
                found = True
                break
            counter += 1
            curNode = curNode.next
        if found == True:
            return counter
        else:
            return -1
    def get_element_by_index(self, index):
        total_len = self.len()
        if index >= total_len or index < 0:
            raise Exception("Linked list index is out of range")
            return
        if index == 0:
            return self.head.data
        curNode = self.head
        counter = 0
        data = -1
        while curNode.next:
            if counter == index - 1:
                data = curNode.next.data
                break
            counter += 1
            curNode = curNode.next
        return data
    def insert_after_value(self, value_after, data_to_insert):
        find_item = self.get_element_position(value_after)
        if find_item == -1:
            raise Exception("Value_after does not exist")
        else:
            curNode = self.head
            while curNode.next:
                if curNode.next.data == value_after:
                    if curNode.next.next:
                        curNode = Node(data_to_insert, curNode.next.next)
                        break
                    else:
                        curNode = Node(data_to_insert, None)
                curNode = curNode.next
    def remove_by_value(self, data):
        find_item = self.get_element_position(data)
        if find_item == -1:
            raise Exception("Value_after does not exist")
        else:
            if self.head.data == data:
                self.remove_element_at(0)
                return
            data_posit = self.get_element_position(data)
            self.remove_element_at(data_posit)
            return
    def insert_at_index(self, index, value):
        total_len = self.len()
        if index >= total_len or index < 0:
            raise Exception("Linked list index out of range")
        counter = 0
        if index == 0:
            self.head = Node(value, self.head)
            return
        curNode = self.head
        while curNode.next:
            if counter == index - 1:
                node = Node(value, curNode.next)
                curNode.next = node
                break
            curNode = curNode.next
            counter += 1
    def flush(self):
        self.head = None
    def print_forward(self):
        curNode = self.head
        result = ""
        if curNode is None:
            print("->")
            return
        while True:
            result += f"{curNode.data}->"
            if curNode.next is None:
                break
            curNode = curNode.next
        print(result)
    def print_backward(self):
        curNode = self.head
        last_node = self.get_last_item_node()
        result = ""
        if curNode is None:
            print("<-")
            return
        while last_node.prev:
            result += f"<-{last_node.data}"
            if last_node.prev is None:
                break
            last_node = last_node.prev
        print(result)
```
