### Numerical Data Plotting

```python
import seaborn as sns
import pandas as pd
import matplotlib as plt
```

```python
%matplotlib inline
```

```python
sns.set_style("whitegrid")
```

```python
tips = sns.load_dataset("tips")
tips.head(50)
```

```python
sns.relplot(x="total_bill", y="tip", data=tips, hue="smoker")
```

```python
sns.relplot(x="total_bill", y="tip", data=tips, hue="time")
```

```python
sns.relplot(x="total_bill", y="tip", data=tips, hue="smoker", style='time')
```

```python
sns.relplot(x="total_bill", y="tip", data=tips, hue="size")
```

```python
sns.relplot(x="total_bill", y="tip", data=tips, hue="smoker", style='size', palette="ch:r=-0.98, l=0.95")
```

```python
sns.relplot(x="total_bill", y="tip", data=tips, hue="smoker", size='size', sizes=(10, 200), palette="ch:r=-0.98, l=0.95")
```

```python
import numpy as np
df = pd.DataFrame(dict(time=np.arange(400), value=np.random.randn(400).cumsum()))
sns.relplot(df, kind='line', x='time', y='value')
```

```python
df2 = pd.DataFrame(np.random.rand(500, 2).cumsum(axis=0), columns=["time", "value"])
sns.relplot(df2, x='time', y='value', kind='line',)
```

```python
fmri = sns.load_dataset("fmri")
sns.relplot(fmri, kind='line', x='timepoint', y='signal')
```

```python
fmri = sns.load_dataset("fmri")
sns.relplot(fmri, kind='line', x='timepoint', y='signal', errorbar=('ci', False))
```

```python
sns.relplot(fmri, kind='line', x='timepoint', y='signal', estimator=None)
```

```python
sns.relplot(fmri, kind='line', x='timepoint', y='signal', hue='event')
```

```python
sns.relplot(fmri, kind='line', x='timepoint', y='signal',style='event', markers=True, hue='event')
```

```python
sns.relplot(x='timepoint', y="signal", hue='region',units='subject', estimator=None, kind="line", data=fmri.query('event == "stim"'))
```

```python
dots = sns.load_dataset("dots").query("align=='dots'")
dots.head()
```

```python
sns.relplot(x='time', y='firing_rate', style='choice', kind='line', data=dots, hue='coherence')
```

```python
df = pd.DataFrame(dict(time=pd.date_range('2019-06-02', periods = 500), value=np.random.randn(500).cumsum(axis=0)))
chart  = sns.relplot(df, x='time', y='value', kind='line')
chart.fig.autofmt_xdate()
```

## catplot

```python
sns.catplot(x='day', y='total_bill', hue='day', data=tips)
```

```python
sns.catplot(x='day', y='total_bill', hue='day', data=tips, jitter=False)
```

```python
sns.catplot(x='day', y='total_bill', hue='day', data=tips,  kind="swarm")
```

```python
sns.catplot(x='day', y='total_bill', data=tips, hue='size', kind="swarm")
```

```python
sns.catplot(x='smoker', y='tip', hue='smoker', data=tips, order=['No', "Yes"])
```

```python
sns.catplot(x='smoker', kind='swarm', y='tip', hue='smoker', data=tips, order=['No', "Yes"])
```

## Box plot

```python
sns.catplot(x='day', y='total_bill', kind='box', data=tips)
```

```python
sns.catplot(x='day', y='total_bill', kind='box', hue='time', data=tips)
```

```python
sns.catplot(x='day', y='total_bill', kind='box', hue='sex', data=tips)
```

```python
sns.catplot(x='day', y='total_bill', kind='box', hue='sex', dodge=False, data=tips)
```

```python
diamonds= sns.load_dataset("diamonds")
```

```python
sns.catplot(x='color', y='price', kind='boxen', data=diamonds.sort_values('color'))
```

```python
sns.catplot(x='color', y='price', kind='violin', data=diamonds.sort_values('color'))
```

```python
sns.catplot(x='total_bill', y='day', kind='violin', data=tips, hue="sex", split=True, inner='stick')
```

```python
g = sns.catplot(x='day', y='total_bill', kind='violin', inner=None, data=tips)
sns.swarmplot(x='day', y='total_bill', color='k', size=3, data=tips, ax=g.ax)
```

## Visualizing Distribution of data
- Distplot
- Kdeplot
- Jointplot
- Rugplot

```python
x= np.random.randn(100)
sns.set_style("darkgrid")
sns.histplot((x), kde=True)
```

```python
sns.histplot(x, element='step', fill=False)
```

```python
sns.jointplot(x="total_bill", y="tip", data=tips)
```

```python
sns.jointplot(x="total_bill", y="tip", data=tips, kind='hex')
```

```python
sns.jointplot(x="total_bill", y="tip", data=tips, kind='kde')
```

```python
sns.regplot(x='total_bill', y='tip', data=tips)
```

```python
sns.lmplot(x='size', y='tip', data=tips, x_jitter=0.1)
```

```python
sns.lmplot(x='size', y='tip', data=tips, x_estimator=np.median)
```
