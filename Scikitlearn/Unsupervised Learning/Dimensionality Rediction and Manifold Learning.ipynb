<h1>Dimensioanlity Reduction and Manifold Learning</h1>

In machine learning problems, we often encounter datasets with a very large number of dimensions (features or columns). Dimensionality reduction techniques are used to reduce the number of dimensions or features within the data to manageable or convenient number.
Applications of dimensionality reduction:
1. Reducing the size of data without loss of information
2. Training machine learning models efficiently
3. Visualizing high-dimensional data in 2/3 dimensions

<h1>Principal Component Analysis (PCA)</h1>
Principal component is a dimensionality reduction technique that uses linear projection of data to reducer their dimensions, while attempting to maximize the variance of data in the projection

Example: PCA Convert from 2D to 1D



```python
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
sns.set_style("darkgrid")
%matplotlib inline
plt.rcParams["figure.figsize"] = (10, 11)
plt.rcParams["font.size"] = 11
```

```python
iris_df = sns.load_dataset("iris")
numeric_cols = iris_df.select_dtypes(np.number).columns.tolist()
```

```python
from sklearn.decomposition import PCA
```

```python
?PCA
```

```python
pca = PCA(n_components=2)
```

```python
pca.fit(iris_df[numeric_cols])
```

```python
transformed = pca.transform(iris_df[numeric_cols])
```

```python
transformed
```

```python
iris_df[numeric_cols]
```

```python
pca.components_
```

```python
sns.scatterplot(x=transformed[:,0], y=transformed[:, 1], hue=iris_df.species)
```

### t-Distributed Stochastic Neighbor Embedding (t-SNE)
Manifold learning is an approach to non-linear dimensionality reduction. Algorithms for this task are based on the idea that the dimensionality of many datasets is only artificially high. Scikit-learn provides many algorithm for manifold learning. A commonly-used manifold learning technique is t-Distributed stochastic neighbor embedding or t-SNE, used to visualize high dimensional data in one, two or three dimensions.
